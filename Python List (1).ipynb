### Question 1: Write a Python program to count the number of elements in a list.


- `sample input : `
    - `var = [10,11,12,13]`
- `sample output :`
    - `result : 4`

```python
var = [10,11,12,13]

element_count = 0

for element in var:
    element_count += 1
print(f"The number of elements in the list is: {element_count}")
```

### Question 2: Write a Python program to find the maximum and minimum values in a given list.

- `sample input : `
    - `var = [10,11,12,13,14,15]`
- `sample output :`
    - `result : max = 15 , min = 10`

```python
mylist = [10, 11, 12, 13, 14, 15]
maxvalue = mylist[0]
minvalue = mylist[0]

for number in mylist:
    if number > maxvalue:
        maxvalue = number
    if number < minvalue:
        minvalue = number
print(f"result:min={minvalue} and max={maxvalue}")


```

### Question 3: Write a Python program to remove the first occurrence of a specified element from a list.

- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : [1,3,4,2,5]`
    

```python
var = [1, 2, 3, 4, 2, 5]
elementtoremove = 2
indextoremove = -1
for i in range(len(var)):
    if var[i] == elementtoremove:
        indextoremove = i
        break
if indextoremove != -1:
    var = var[:indextoremove] + var[indextoremove + 1:]
print(f"result : {var}")
```

### Question 4: Write a Python program to find the average of all the elements in a list.

### Question 7: Write a Python program to find which number is palindrome in below given list .

- `sample input : `
    - `var = [134,121,155]`
- `sample output :`
    - `result : 134 is not a palindome , 121 is a palindrome `

```python
var= [134, 125, 155]
for number in var:
    if str(number) == str(number)[::-1]:
        print(f"{number} is a palindrome.,")
    else:
        print(f"{number} is not a palindrome.")
```

### Question 8: Write a Python program to find the common elements between two lists.

- `sample input : `
    - `var1 = [1,2,3,4,10]`
    - `var2 = [10,11,12,14]`
- `sample output :`
    - `result : 10`

```python
var1 = [1, 2, 3, 4, 10]
var2 = [10, 11, 12, 14]
common = set(var1) & set(var2)
print(f"result: {list(common)}")
```
