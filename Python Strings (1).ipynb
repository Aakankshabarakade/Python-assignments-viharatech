##Question** 1: Write a Python program to count the number of characters in a string.
# Sample Input 1

     "Hello, World!"

     count_characters(input_string1)

# Sample Input 2

    "Python is fun."

       count_characters(input_string2)

```python
string1 = "hello world"
string2 = "python is fun"
count_string1 = 0
for char in string1:
    if char.strip():
        count_string1 += 1
count_string2 = 0
for char in string2:
    if char.strip():
        count_string2 += 1
print(f"Number of characters in string1: {count_string1}")
print(f"Number of characters in string2: {count_string2}")
```

Question 3: Write a Python program to check if a string is a palindrome




# Sample Input 1

input_string1 =

       "racecar"

output1 =

    is_palindrome(input_string1)

# Sample Input 2

input_string2 =

    "Hello, World!"

output2 =

    is_palindrome(input_string2)

```python
str1 = "racecar"
str2 = "hello world"
def is_palindrome(string):
    length = len(string)
    for i in range(length // 2):
        if string[i] != string[length - i - 1]:
            return False
    return True
if is_palindrome(str1):
    print(f"'{str1}' is a palindrome.")
else:
    print(f"'{str1}' is not a palindrome.")
if is_palindrome(str2):
    print(f"'{str2}' is a palindrome.")
else:
    print(f"'{str2}' is not a palindrome.")
```

Question 4: Write a Python program to find the most common character in a string.






# Sample Input 1

input_string1 =

    "hello world"

output1 =

    most_common_character(input_string1)

# Sample Input 2

input_string2 =

    "python programming"

output2 =

    most_common_character(input_string2)


```python
str1 = "hello world"
str2 = "python programming"
combined_string = str1 + str2
char_count = {}
for char in combined_string:
    char_count[char] = char_count.get(char, 0) 
most_common_char = max(char_count, key=char_count.get)
print(f"The most common character in both strings is: {most_common_char}")
```

```python
input_string = "Hello, World!"
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    return ''.join(char for char in input_string if char not in vowels)
result = remove_vowels(input_string)
print("String after removing vowels:", result)
```

Question 9: Write a Python program to find the frequency of each character in a string.




# Sample Input1

input_string1 =

         "Hello, World!"

# Sample Output

output1 =
        character_frequency(input_string1)

# Sample Input2

input_string2 =

        "Python is great"

# Sample Output

output2 =     

      character_frequency(input_string2)


```python
str1 = "hello world"
str2 = "python is great"
def count_characters(input_string):
    char_count = {}
    for char in input_string:
        if char != ' ':
            char_count[char] = char_count.get(char, 0) + 1
    return char_count
char_count_str1 = count_characters(str1)
char_count_str2 = count_characters(str2)
print("Frequency of characters in str1:")
for char, count in char_count_str1.items():
    print(f"{char}: {count}")
print("\nFrequency of characters in str2:")
for char, count in char_count_str2.items():
    print(f"{char}: {count}")
```

Question 10: write a python programme to find the sum of all the even characters


# Sample Input1

input_string1 =

     "Hello, World!"

# Sample Output

output1 =

     sum_of_even_characters(input_string1)

# Sample Input2

input_string2 =

        "Python is great"

# Sample Output

output2 =

        sum_of_even_characters(input_string2)


```python
str1 = "hello world"
str2 = "python is fun"
def sum_even_characters(input_string):
    total_sum = 0
    for i in range(len(input_string)):
        if i % 2 == 0:
            total_sum += ord(input_string[i])
    return total_sum
sum_str1 = sum_even_characters(str1)
sum_str2 = sum_even_characters(str2)
print("Sum of even characters:", sum_str1)
print("Sum of even characters:", sum_str2)
```
